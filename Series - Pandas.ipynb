```python
fruits = ["orange", "apple","coconut","berry","pineapple"]

days = ["monday", "tuesday", "wednesday", "thursday", "friday"]

#hold shift and press tab to see the help

pd.Series(fruits, days)
```

```python
#or

pd.Series(data = fruits, index= days)
```

```python
pd.Series(data = fruits, index= days, name = 'Fruit of the day')
```

```python
fruits = ["orange", "apple","coconut","berry","pineapple", "nuts"]

days = ["monday", "tuesday", "wednesday", "thursday", "friday", "monday"]

pd.Series(data = fruits, index= days)
```
