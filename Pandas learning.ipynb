```python
import pandas
```

```python
df1 = pandas.read_csv("supermarkets.csv")
df1
```

```python
df1["Address"] = df1["Address"] + ' ,' + df1['City'] + ' ,'+ df1['State'] +' ,' + df1['Country']
df1
```

```python
# can create a new column where we store the string df["Coordinates"]
```

```python
df1["Coordinates"] = df1["Address"].apply(nom.geocode) #nom.geocode came from above & no brackets because the apply does it
df1
```

```python
print (df1.Coordinates[0])
```

```python
print (df1.ix[3,"Coordinates"])
```

```python
df1.Coordinates[0]
```

```python
df1.Coordinates[0].latitude
```

```python
df1.Coordinates[0].longitude
```

```python
df1["Latitude"] = df1["Coordinates"].apply(lambda x : x.latitude if x !=None else "No Coords") # x is temporary variable
df1
```

```python
df1["Longitude"] = df1["Coordinates"].apply(lambda y : y.longitude if y !=None else "No Coords") # y is temporary variable
df1
```
